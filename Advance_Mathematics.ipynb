```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/data.csv", encoding='latin1')
print(df.head())
```

```python
x = df['no2'].dropna().values
```

```python
roll_no = 102316094
ar = 0.05*(roll_no % 7)
br = 0.3*((roll_no %5) +1)

z = x + ar*np.sin(br*x)
```

```python
mu = np.mean(z)
sigma = np.var(z)
lambda_ = 1/ (2*sigma)
c = np.sqrt(lambda_ / np.pi)
```

```python
print(f"Estimated parameters :\n mu = {mu}\n lambda = {lambda_}\n c = {c}")
```
